# Cumalative Sales (cumsum)
---

```python
cumsum  = np.cumsum(sales_data[:,1:],axis=1)
print("Cumalative sales By Year\n",cumsum)
```

# Graph Of Cumsum

```python
plt.figure(figsize=(12,8))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Sale Of Restaurent")
plt.xlabel("Years")
plt.ylabel("Sale Value")
plt.grid(True)
plt.show()
```
