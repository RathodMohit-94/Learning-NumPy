```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
array1 = np.array([[2,3,4,1],[3,5,24,234]])
array2 = np.zeros((5,5))
array3 = np.random.rand(3,5)

```

```python
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)
```

```python
load_arr1 = np.load("array1.npy")
print(load_arr1)

load_arr2 = np.load("array2.npy")
print(load_arr2)

load_arr3 = np.load("array3.npy")
print(load_arr3)
```

```python
try:
    pic1 = np.load("numpy-logo.npy")

    #Display Image
    plt.figure(figsize=(10,10))
    plt.subplot(121)
    plt.title("LOGO")
    plt.imshow(pic1)
    plt.grid(False)

    darkpic = 1 - pic1
    plt.subplot(122)
    plt.title("Dark LOGO")
    plt.imshow(darkpic)
    plt.grid(False)

    

except FileExistsError:
    print("File Not  Found...")
```
